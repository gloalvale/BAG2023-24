```python
# Versión Ana Belén
def get_mult(seq):
    mm = {'A':0, 'C':0, 'G':0, 'T':0}
    for n in seq:
        mm[n] += 1
    for n in mm:
        mm[n] /= len(seq)
    return (mm)
```

```python
def get_transition_matrix(dimeros):
    alphabet = ['A', 'B', 'C', 'D']
    # para cada letra...
    for nucleotido1 in alphabet:
        #inicializando sumatorio a 0
        suma_nucleotido1 = 0
        #para acumular el conteo que aparece en dímeros.
        for nucleotido2 in alphabet:
            key = nucleotido1 + nucleotido2
            suma_nucleotido1 += dimeros[key]
        # Uso este sumatorio para dividir la freq abs
        for nucleotido2 in alphabet:
            #recorra los dimeros de nuevo
            key = nucleotido1 + nucleotido2
            dimeros[key] /= suma_nucleotido1
    return(dimeros)
```
