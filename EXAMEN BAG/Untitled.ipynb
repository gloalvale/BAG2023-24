```python
def limpiar_texto3(dataset):
    dataset = dataset.astype(str)
    for indice, fila in dataset.iterrows():
        texto = fila['text']

        
        tknzr = TweetTokenizer()
        stop_words = set(stopwords.words('english'))

        texto_procesado = re.sub(r'(@\w+|#\w+|http\S+)', '', texto)
        texto_minus = texto_procesado.lower()
        texto_tokens = tknzr.tokenize(texto_minus)

        # Inicializar la lista de stem_words para cada fila
        filtered = []

        for word in texto_tokens:
            filtered = [word for word in texto_tokens if not word in stop_words]
            stemmer = SnowballStemmer('english')
            stem_words = [stemmer.stem(word) for word in filtered]

        final_text = ' '.join(stem_words)
        dataset.at[indice, 'text'] = final_text

    return dataset
```
