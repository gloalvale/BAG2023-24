```python
import re
from nltk.tokenize import TweetTokenizer
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
import pandas as pd

def limpiar_texto(dataset):
    dataset = dataset.astype(str)
    # Descargar las stop words si no están descargadas
    import nltk
    nltk.download('stopwords')
    
    # Crear el tokenizador de tweets
    tokenizer = TweetTokenizer()

    # Obtener las stop words en inglés
    stop_words = set(stopwords.words('english'))

    # Crear el lematizador (stemmer)
    stemmer = SnowballStemmer('english')

    # Iterar sobre cada fila del dataset
    for index, row in dataset.iterrows():
        # Obtener el texto de la fila
        texto = row['text']  # Asegúrate de que la columna en tu DataFrame se llame 'text'

        # Eliminar menciones, hashtags y URLs
        texto = re.sub(r'@\S+|#[^\s]+|http[s]?://\S+', '', texto)

        # Convertir el texto a minúsculas
        texto = texto.lower()

        # Tokenizar el texto
        palabras = tokenizer.tokenize(texto)

        # Eliminar stop words y aplicar lematización
        palabras_filtradas = [stemmer.stem(p) for p in palabras if p not in stop_words]

        # Convertir la lista de palabras filtradas a una cadena de texto
        texto_filtrado = ' '.join(palabras_filtradas)

        # Actualizar el dataset con el texto filtrado
        dataset.at[index, 'text'] = texto_filtrado

    return dataset

# Crear DataFrame con el ejemplo de texto
data = {'text': ["sooo sad miss san diego ! ! !",
                 "boss bulli ...",
                 "interview ! leav alon",
                 "son * * * , ` put relea alreadi bough"]}

df = pd.DataFrame(data)

# Aplicar la función para limpiar el texto en el dataset
df_limpiado = limpiar_texto(df)

# Imprimir el DataFrame después de la limpieza
print(df_limpiado)
```

```python
import pandas as pd
import re
from nltk.tokenize import TweetTokenizer
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

def limpiar_textoaa(dataset):
    # Descargar las stop words si no se han descargado previamente
    import nltk
    nltk.download('stopwords')
    dataset['text'] = dataset['text'].astype(str)
    
    # Inicializar el tokenizador de tweets y el stemmer
    tokenizer = TweetTokenizer()
    stemmer = SnowballStemmer('english')
    
    # Obtener la lista de stop words en inglés
    stop_words = set(stopwords.words('english'))
    
    # Iterar sobre cada fila del DataFrame
    for index, row in dataset.iterrows():
        # Obtener el texto de la fila
        texto = row['text']
        
        # Eliminar menciones, hashtags y URLs
        texto = re.sub(r'@[^\s]+|#[^\s]+|http[^\s]+|www[^\s]+', '', texto)
        
        # Convertir a minúsculas
        texto = texto.lower()
        
        # Tokenización
        palabras = tokenizer.tokenize(texto)
        
        # Eliminar stop words y aplicar lematización (stemming)
        palabras = [stemmer.stem(palabra) for palabra in palabras if palabra not in stop_words]
        
        # Unir las palabras en una cadena separada por espacios
        texto_final = ' '.join(palabras)
        
        # Reemplazar el texto original en el DataFrame
        data.at[index, 'text'] = texto_final
    
    return dataset

# Cargar datos
data = pd.read_csv('Tweets.csv', usecols=["text"])
d = data[1:5]

# Aplicar la función de limpieza
limpiar_texto(d)


```
